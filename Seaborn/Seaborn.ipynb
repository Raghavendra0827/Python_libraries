```python
import pandas as pd
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
sns.pairplot(data = df, hue = 'brand')
```

```python
sns.distplot(df['mpg'] , color ='red', bins =5)
```

```python
# Question 1
# Create a scatter plot for the following data
# hp and mpg
# hp and time to 60
# hp and weightlbs
# hp and cylinders
```

```python
fig, axes = plt.subplots(2,2, figsize = (20,10))
plt.subplot(2,2,1)
sns.scatterplot(x='mpg', y = 'hp', data = df)
plt.title('MPG to HP data')
plt.xlabel("Miles per Gallon")
plt.ylabel("Horsepower")

plt.subplot(2,2,2)
sns.scatterplot(x='time-to-60', y = 'hp', data = df)
plt.title('MPG to Time to 60 data')
plt.xlabel("Miles per Gallon")
plt.ylabel("Time-to-60")

plt.subplot(2,2,3)
sns.scatterplot(x='weightlbs', y = 'hp', data = df)
plt.title('MPG to Time to 60 data')
plt.xlabel("Miles per Gallon")
plt.ylabel("Time-to-60")

plt.subplot(2,2,4)
sns.scatterplot(x='cylinders', y = 'hp', data = df)
plt.title('MPG to Time to 60 data')
plt.xlabel("Miles per Gallon")
plt.ylabel("Time-to-60")
```

```python
# Question 2
# Create a bar plot for the following data
# cylinders and hp,
# cylinders and mpg
# cylinders and weight
# year and time-to-60
# year and mpg
# year and hp


fig,axes=plt.subplots(3,2,figsize = (20,10))
plt.subplot(3,2,1)
sns.barplot(x='cylinders',y='hp',data = df)
plt.title('Cylinders to HP Data')
plt.xlabel('Cyclinders')
plt.xlabel('Horse power')

plt.subplot(3,2,2)
sns.barplot(x='cylinders',y='mpg',data = df)
plt.title('Cylinders to Mpg Data')
plt.xlabel('Cylinders')
plt.xlabel('Miles per gallon')


plt.subplot(3,2,3)
sns.barplot(x='cylinders',y='weightlbs',data = df)
plt.title('Cylinders to Weight Data')
plt.ylabel('Weight in lbs')
plt.xlabel('Cylinders')


plt.subplot(3,2,4)
sns.barplot(x='year',y='time-to-60',data = df)
plt.title('Year to Time-to-60 Data')
plt.xlabel('Year')
plt.xlabel('Time-to-60')


plt.subplot(3,2,5)
sns.barplot(x='year',y='mpg',data = df)
plt.title('Year to MPGData')
plt.xlabel('Year')
plt.xlabel('MPG')

plt.subplot(3,2,6)
sns.barplot(x='year',y='hp',data = df)
plt.title('Year to HP Data')
plt.xlabel('Year')
plt.xlabel('Horse power')

```
