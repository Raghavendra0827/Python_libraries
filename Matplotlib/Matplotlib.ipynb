```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0,10,0.1)
y = 2*x+5

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,1)
y = 2*x+5

fig = plt.figure(figsize=(10,10))
plt.plot(x,y, linewidth = 1, linestyle = '-', color = 'r', alpha = 0.9, marker = 'H')
plt.title("Line Plot Demo")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Line1'], loc = 'upper right')
plt.grid(True)
```

```python
x = np.arange(0,10,1)
y1 = 2*x+5
y2 = 3*x+10

plt.subplot(1,2,1)
plt.plot(x,y1, linewidth = '4')
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')
```

```python
data = {'apples':20,'mangoes':30,'lemon':25}
names = list(data.keys())
values = list(data.values())

plt.barh(names, values, color = 'orange')
plt.title('Bar chart')
plt.xlabel('Fruits count')
plt.ylabel('Fruits')
plt.legend(['Bars'], loc = 'best')
```

```python
a = np.arange(10,60,10)
b = [5,3,2,6,1]
c = [1,2,3,4,6]

plt.scatter(a,b, c='g', s=200, marker = 's', alpha = 0.6)
plt.scatter(a,c, c = 'r', s =400, marker = 'H', alpha = 0.7)

plt.title('Scatterplot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend(['Scatter1', 'Scatter2'], loc = 'best')
```
