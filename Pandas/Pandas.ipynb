# Merge ,join and concatenate
##1.how to perform merge operation using pands?
##2.how to perform join operation?
##2.how to perform concantenate operation?

```python
import pandas as pd
player=['narasimha','kempa','raghu']
point=[10,9,8]
title=['cricket','kabbadi','kho-kho']
df=pd.DataFrame({'player':player,'points':point,'title':title})
df
```

```python
import pandas as pd
player=['narasimha','kempa','manu']
power=['punch','kick','elbow']
title=['game1','game2','game3']
df1=pd.DataFrame({'player':player,'power':power,'title':title})
df1
```

```python
# inner merge
df1.merge(df,on='player',how='inner')
```

```python
# outer merge / merge
df1.merge(df,on='player',how='outer')
```

```python
# left merge
df1.merge(df,on='player',how='left') ## which gives all the data from left table that is df1
```

```python
# right merge
df1.merge(df,on='player',how='left') ## which gives all the data from right table that is df1
```

# to make join operation we need index value must be same

```python
import pandas as pd
player=['narasimha','kempa','raghu']
point=[10,9,8]
title=['cricket','kabbadi','kho-kho']
df2=pd.DataFrame({'player':player,'points':point,'title':title},index=[1,2,3])
df2
```

```python
import pandas as pd
player=['narasimha','kempa','manu']
power=['punch','kick','elbow']
title=['game1','game2','game3']
df3=pd.DataFrame({'player':player,'power':power,'title':title},index=[2,3,4])
df3
```

```python
#inner join
df2.join(df3, how='inner', lsuffix='_x', rsuffix='_y')## we must provide suffix because two table has same attribute name
```
